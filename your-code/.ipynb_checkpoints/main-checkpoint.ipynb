````python
div0=divs[0].find_all('a')
div0
#We have name and alias:
````

````python
for item in div0:
    print(item.text.strip()) #without spaces
````

````python
#Now for all the list:
````

````python
#I store the users info in a dictionary:
users=[]
for div in divs:
    for username,alias in zip(div.select('.h3 > a'),div.select('.f4 > a')):  #soup.select('tag1 > tag2') tag2 is a child of tag1
        #print(item)
        #print(username.text.strip(),alias.text.strip())
        users.append({'username' : username.text.strip(), 'alias' : alias.text.strip()})
users
````

````python
#As requested, users info displayed on a list for final output:
list_result=[]
for user in users:
    list_result.append(user['username'] + ' ('+ user['alias']  + ')')
list_result
    
````

#### 1.1. Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
# your code here
response = requests.get(url)
response
````

````python
# turning the response into a beautiful soup object
soup = BeautifulSoup(response.content)
````

````python
articles = soup.find_all('article', attrs= {'class' : 'Box-row'})
articles
````

````python
#I store the users info in a dictionary:
users=[]
for article in articles:
    for username in article.select('.h3 > a'):  #soup.select('tag1 > tag2') tag2 is a child of tag1
        print(username.text.strip().replace('\n', ' ').replace('       ', ''))

users
````
